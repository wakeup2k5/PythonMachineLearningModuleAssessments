# Preprocessing
Perform any needed pre-processing on the chosen features, including:

Dealing with NaN values

Note: You need to output the result of your pre-processing to an output CSV called “data_refined.csv.”

```python
# Encoding the Diagnosis column (Benign = 0, Malignant = 1)

from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()

encoder.fit(df['diagnosis'])

df['diagnosis'] = encoder.transform(df['diagnosis'])

df
```

```python
# Scaling the features
from sklearn.preprocessing import StandardScaler
import numpy as np

scaler = StandardScaler()

y = df.iloc[:,0].values
x_train = df.iloc[:,1:20].values

x_scaled = scaler.fit_transform(x_train)

np.savetxt('data_refined.csv', x_scaled, delimiter=',', fmt='%s')
x_scaled

```

```python
# pair plots

import matplotlib.pyplot as plt
import seaborn as sns

subset_features = ["radius_mean","texture_mean","perimeter_mean","area_mean","smoothness_mean"]

# Create a pair plot with the selected features and color by 'target_label'
sns.pairplot(df[subset_features + ['diagnosis']], hue='diagnosis', palette='coolwarm')

# Display the plot
plt.show()
```

```python
# heatmap

sns.heatmap(data = df.corr())
```

```python
# box plot

sns.boxplot(x='diagnosis', y='radius_mean', data=df, palette='coolwarm')

# Set the title and labels
plt.title('Box Plot of Mean Radius by Tumor Type')
plt.xlabel('Tumor Type (Benign vs Malignant)')
plt.ylabel('Mean Radius')
```
